```python
import torch
import torch.nn as nn
from typing import Callable
from torch.nn.functional import relu
```

```python
# Exercise 4.1
class MLP(nn.Module):
    def __init__(self, hidden_channels,out_channels):
        super(MLP, self).__init__()
        self.h=hidden_channels
        self.o=out_channels
        
        self.f1 = nn.Linear(3, self.h)
        self.f2=nn.Linear(self.h,self.h)
        self.f3=nn.Linear(self.h,self.o)
         # <YOUR CODE GOES HERE
   
    def forward(self, x):
        # <YOUR CODE GOES HERE
        print(x.shape,"x shape")
        a=self.f1(x)
        print(a.shape," ashape")
        b = relu(a)
        f2_x= self.f2(b)
        print(f2_x.shape," f2_x shape")
        c=relu(f2_x)
        print(c.shape,"c shape")
        d=self.f3(c)
        print(d.shape,"d shape")
        return a,b,c,d

H = torch.randint(5, 10, (1,)).item()
C = torch.randint(2, 5, (1,)).item()
net = MLP(H,C)
# print(net.h)
x = torch.rand(16,3)
net.forward(x)
```
